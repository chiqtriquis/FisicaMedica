```python
orbitales = [
    ('1s', 2), ('2s', 2), ('2p', 6), ('3s', 2), ('3p', 6), ('4s', 2),
    ('3d', 10), ('4p', 6), ('5s', 2), ('4d', 10), ('5p', 6), ('6s', 2),
    ('4f', 14), ('5d', 10), ('6p', 6), ('7s', 2), ('5f', 14), ('6d', 10), ('7p', 6)
]
excepciones = {
    24: "1s2 2s2 2p6 3s2 3p6 3d5 4s1",  # Cromo (Cr)
    29: "1s2 2s2 2p6 3s2 3p6 3d10 4s1",  # Cobre (Cu)
    41: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d4 5s1",  # Niobio (Nb)
    42: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d5 5s1",  # Molibdeno (Mo)
    44: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d7 5s1",  # Rutenio (Ru)
    45: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d8 5s1",  # Rodio (Rh)
    46: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10",     # Paladio (Pd)
    47: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s1",  # Plata (Ag)
    57: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 4f1 5d1 6s2",  # Lantano (La)
    58: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 4f1 5d1 6s2",  # Cerio (Ce)
    64: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 4f7 5d1 6s2",  # Gadolinio (Gd)
    79: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 4f14 5d10 6s1", # Oro (Au)
    89: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 5d1 6s2",       # Actinio (Ac)
    90: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 5f2 6d2 7s2",   # Torio (Th)
    91: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 5f2 6d3 7s2",   # Protactinio (Pa)
    92: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 5f3 6d3 7s2",   # Uranio (U)
    93: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 5f4 6d3 7s2",   # Neptunio (Np)
    96: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 5f7 6d1 7s2",   # Curio (Cm)
    103: "1s2 2s2 2p6 3s2 3p6 3d10 4s2 4p6 4d10 5s2 5p6 5f14 6d1 7s2"  # Lawrencio (Lr)
}

numero = int(input("Ingresa un número atómico Z tal que 1 ≤ Z ≤ 127: "))
if numero in df['Z'].values:
    simbolo = df[df['Z'] == numero]['Símbolo'].values[0]
    nombre = df[df['Z'] == numero]['Nombre'].values[0]
    print(f"Elemento: {simbolo} ({nombre})")
else:
    print(f"El número ingresado no corresponde a ningún elemento.")


def configuracion_electronica(numero):
    if numero in excepciones:
        return excepciones[numero]

    electrones_restantes = numero
    configuracion = []

    for orbital, capacidad in orbitales:
        if electrones_restantes > capacidad:
            configuracion.append(f"{orbital}{capacidad}")
            electrones_restantes -= capacidad
        else:
            configuracion.append(f"{orbital}{electrones_restantes}")
            break

    return ' '.join(configuracion)

# Función para calcular el periodo y el grupo
def periodo_y_grupo(numero):
    if numero in excepciones:
        config = excepciones[numero]
    else:
        config = configuracion_electronica(numero)

    niveles = [int(orbital[0]) for orbital in config.split()]
    nivel_energia_maximo = max(niveles)

    # Contar los electrones en el último nivel de energía para determinar el grupo
    electrones_s = electrones_p = electrones_d = electrones_f = 0

    for orbital in config.split():
        nivel = int(orbital[0])
        tipo = orbital[1]
        num_electrones = int(orbital[2:])

        if nivel == nivel_energia_maximo:
            if tipo == 's':
                electrones_s = num_electrones
            elif tipo == 'p':
                electrones_p = num_electrones
            elif tipo == 'd':
                electrones_d = num_electrones
            elif tipo == 'f':
                electrones_f = num_electrones

    # Determinar el grupo según el tipo de orbital y configuración
    if electrones_s > 0 and electrones_p == 0 and electrones_d == 0 and electrones_f == 0:
        grupo = 1 if electrones_s == 1 else 2
    elif electrones_p > 0:
        grupo = 13 + electrones_p
    elif electrones_d > 0:
        grupo = 3 + electrones_d
    elif electrones_f > 0:
        grupo = 3

    return nivel_energia_maximo, grupo

# Función para calcular los números cuánticos del último electrón
def numeros_cuanticos_ultimo_electron(config):
    ultimo_orbital = config.split()[-1]
    nivel_energia = int(ultimo_orbital[0])
    tipo_orbital = ultimo_orbital[1]
    num_electrones = int(ultimo_orbital[2:])

    if tipo_orbital == 's':
        l = 0
    elif tipo_orbital == 'p':
        l = 1
    elif tipo_orbital == 'd':
        l = 2
    elif tipo_orbital == 'f':
        l = 3

    electrones_en_orbital = num_electrones
    m_l = (electrones_en_orbital - 1) % (2 * l + 1) - l
    ms = (-1) ** (electrones_en_orbital - 1) * 0.5

    return (nivel_energia, l, m_l, ms)

# Configuración electrónica
configuracion=configuracion_electronica(numero)
print(f"La configuración electrónica es: {configuracion}")
#  periodo y el grupo
periodo, grupo = periodo_y_grupo(numero)
print(f"El periodo es: {periodo}")
print(f"El grupo es: {grupo}")

n, l, m_l, ms = numeros_cuanticos_ultimo_electron(configuracion)
print(f"Números cuánticos : n={n}, l={l}, ml={m_l}, ms={ms}")
```

```python
cterydberg=1.097*(10**7)
NivelInicial=[]
Lyman=[]
Balmer=["nan"]
Paschen=["nan", "nan"]
Brackett=["nan", "nan", "nan"]
Pfund=["nan", "nan", "nan", "nan"]
Humphreys=["nan", "nan", "nan", "nan", "nan"]
def serie_de_Rydberg(nivelinicial):
    for l in range(nivelinicial):
        if (l>0):
            NivelInicial.append(l+1)
    def serie_de_Lyman(nivelinicial):
        for l in range(nivelinicial):
            if (l>0):
                lambda_lyman=(1/((cterydberg*(numero**2))*((1/1**2)-(1/(l+1)**2))))*10**9
                lambda_lyman_rounded = round(lambda_lyman, 5)
                Lyman.append(f"{lambda_lyman_rounded} nm")
        #print(Lyman)
    def serie_de_Balmer(nivelinicial):
        for l in range(nivelinicial):
            if (l>1):
                lambda_balmer=(1/((cterydberg*(numero**2))*((1/2**2)-(1/(l+1)**2))))*10**9
                lambda_balmer_rounded = round(lambda_balmer, 5)
                Balmer.append(lambda_balmer_rounded)
        #print(Balmer)
    def serie_de_Paschen(nivelinicial):
        for l in range(nivelinicial):
            if (l>2):
                lambda_paschen=(1/((cterydberg*(numero**2))*((1/3**2)-(1/(l+1)**2))))*10**9
                lambda_paschen_rounded = round(lambda_paschen, 5)
                Paschen.append(lambda_paschen_rounded)
    def serie_de_Brackett(nivelinicial):
        for l in range(nivelinicial):
            if (l>3):
                lambda_brackett=(1/((cterydberg*(numero**2))*((1/4**2)-(1/(l+1)**2))))*10**9
                lambda_brackett_rounded = round(lambda_brackett, 5)
                Brackett.append(lambda_brackett_rounded)
    def serie_de_Pfund(nivelinicial):
        for l in range(nivelinicial):
            if (l>4):
                lambda_pfund=(1/((cterydberg*(numero**2))*((1/5**2)-(1/(l+1)**2))))*10**9
                lambda_pfund_rounded = round(lambda_pfund, 5)
                Pfund.append(lambda_pfund_rounded)
    def serie_de_Humphreys(nivelinicial):
        for l in range(nivelinicial):
            if (l>5):
                lambda_humphreys=(1/((cterydberg*(numero**2))*((1/6**2)-(1/(l+1)**2))))*10**9
                lambda_humphreys_rounded = round(lambda_humphreys, 5)
                Humphreys.append(lambda_humphreys_rounded)
    serie_de_Lyman(nivelinicial)
    serie_de_Balmer(nivelinicial)
    serie_de_Paschen(nivelinicial)
    serie_de_Brackett(nivelinicial)
    serie_de_Pfund(nivelinicial)
    serie_de_Humphreys(nivelinicial)
nivelinicial = int(input("Ingresa el nivel inicial de energía tal que n0>=6"))
print(f"A continuación se presentan las longitudes de onda correspondientes para Z={numero}, teniendo en cuenta la ecuación de la serie de Rydberg:")
serie_de_Rydberg(nivelinicial)
espectros = pd.DataFrame({
})
espectros['N.Inicial(n0)'] = NivelInicial
espectros['Lyman(nm)(nf=1)'] = Lyman
espectros['Balmer(nm)(nf=2)'] = Balmer
espectros['Paschen(nm)(nf=3)'] = Paschen
espectros['Brackett(nm)(nf=4)'] = Brackett
espectros['Pfund(nm)(nf=5)'] = Pfund
espectros['Humphreys(nm)(nf=6)'] = Humphreys
print(espectros)
```
